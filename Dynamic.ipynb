## Output Dynamic Update 
### 1.1 Plot and Print 
if needed to hold on the last output just do not clear plt

```python
import pylab as plt
from IPython import display
for i in range(10):
    plt.plot(plt.randn(100))
    display.clear_output(wait=True)
    display.display(plt.gcf())
    plt.gca().clear()
    print(i)
```
